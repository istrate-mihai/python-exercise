```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_coords = [1, 2, 3, 4]
y_coords = [5, 6, 5, 7]

plt.plot(x_coords, y_coords)
plt.xlabel('x - Axis')
plt.ylabel('y - Axis')
plt.title('Basic Line Plot')
plt.show()
```

```python
x1 = np.arange(0, 9, 2)
x2 = np.arange(10, 19, 2)
y1 = np.arange(20, 29, 2)
y2 = np.arange(30, 39, 2)

plt.figure(figsize=(9, 5))

plt.subplot(2, 2, 1)
plt.plot(x1, y1, color='red', linestyle='-')
plt.title('Plot 1')


plt.subplot(2, 2, 2)
plt.plot(x1, y2, color='green', linestyle='--')
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x2, y1, color='blue', linestyle='-.')
plt.title('Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x2, y2, color='brown', linestyle='-')
plt.title('Plot 4')
```

```python
#pie chart

labels  = ['A', 'B', 'C', 'D', 'E']
values  = [10, 20, 50, 20, 5]
colors  = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0, 0)

plt.pie(values, explode = explode, labels = labels, colors = colors, autopct = "%1.1f%%", shadow = True)
plt.show()
```
